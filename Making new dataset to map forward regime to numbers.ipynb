# Making new dataset to map forward regime to numbers

```python
import pandas as pd


input_path = ("C:\\Users\\user\\Downloads\\market classification project using TCN on interaday Gold data\\hourly features formula one.csv")

output_path = ("C:\\Users\\user\\Downloads\\market classification project using TCN on interaday Gold data\\hourly features edited ver.csv")
               
df = pd.read_csv(input_path)

# map text labels to integers

label_map = {
    "downward": 0,
    "range": 1,
    "upward": 2
}

df["forward regime"] = df["forward regime"].map(label_map)

# valiate
if df["forward regime"].isnull().any():
    print('error occurred were not mapped correctly:', df["forward regime"].unique())
else:
    print('mapping successful. saving dataset...')


df.to_csv(output_path, index=False)
print('saved:', output_path)
```
